```python
total_missing_values = df.isnull().sum().sum()
print(f'total missing values in the datset: {total_missing_values} ({round((total_missing_values/df.shape[0])*100, 2)}%)')
```

```python
for col in df.columns:   #FOR HANDLING MISSING VALUES. WE USED MEDIAN OF THAT COLUMN.
    if df[col].isnull().sum() > 0:
        df[col].fillna(df[col].median(), inplace=True)
```

HOW MANY OF THE CUSTOMERS RETAIN AND HOW MANY CHURN

```python
counts = df['Churn'].value_counts()
print(counts)
plt.figure(figsize= (5,5))
plt.pie(counts, labels=['Retained', 'Churned'], autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.show()

```

```python
plt.figure(figsize=(8, 5))

# Create separate histograms for male and female customers
sns.histplot(data=df[df['Gender'] == 'Male'], x='SatisfactionScore', color='blue', label='Male')
sns.histplot(data=df[df['Gender'] == 'Female'], x='SatisfactionScore', color='pink', label='Female')

# Set the title and labels
plt.title('Distribution of Satisfaction Score by Gender')
plt.xlabel('Satisfaction Score')
plt.ylabel('number of customers')

# Show the legend
plt.legend()

# Show the plot
plt.show()

```

```python
# Group data by satisfaction score and calculate churn rate (percentage of churned customers)
churn_by_satisfaction = df.groupby('SatisfactionScore')['Churn'].mean()

# Create a bar chart
plt.figure(figsize=(8, 6))
plt.bar(churn_by_satisfaction.index, churn_by_satisfaction.values)

# Customize the plot for clarity
plt.xlabel("Satisfaction Score")
plt.ylabel("Churn Rate (%)")
plt.title("Churn Rate by Satisfaction Score")

# Display the plot
plt.show()
```
