```python
import requests
import pandas as pd

# Replace with your CSV file path
csv_file = "C:/Users/Mali/ai-project-methodology/data/Dataset/ECommerce.csv"

# Open the CSV file in binary mode for reading
with open(csv_file, 'rb') as f:
    csv_data = f.read()  # Read the entire CSV file content

# Set the API URL
url = "http://localhost:5000/predict"

# Send the POST request with the CSV file content
files = {'file': (csv_file, csv_data)}  # Use a dictionary for file upload
headers = {}  # Headers for multipart/form-data (optional)
response = requests.get(url, files=files, headers=headers)

# Check for successful response
if response.status_code == 200:
    data = response.json()
    print(f"Predicted churn: {data}")
else:
    print(f"Error: {response.status_code}")
    print(response.text)
```
